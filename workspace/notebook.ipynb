## 💾 The data

SleepInc has provided you with an anonymized dataset of sleep and lifestyle metrics for 374 individuals. This dataset contains average values for each person calculated over the past six months.

The dataset includes 13 columns covering sleep duration, quality, disorders, exercise, stress, diet, demographics, and other factors related to sleep health. 

| Column | Description |
|---------|----------------------------------------|  
| `Person ID` | An identifier for each individual. |
| `Gender` | The gender of the person (Male/Female). |  
| `Age` | The age of the person in years. |
| `Occupation` | The occupation or profession of the person. |
| `Sleep Duration (hours)` | The average number of hours the person sleeps per day. |
| `Quality of Sleep (scale: 1-10)` | A subjective rating of the quality of sleep, ranging from 1 to 10. |
| `Physical Activity Level (minutes/day)` | The average number of minutes the person engages in physical activity daily. |  
| `Stress Level (scale: 1-10)` | A subjective rating of the stress level experienced by the person, ranging from 1 to 10. |
| `BMI Category` | The BMI category of the person (e.g., Underweight, Normal, Overweight). |
| `Blood Pressure (systolic/diastolic)` | The average blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure. |
| `Heart Rate (bpm)` | The average resting heart rate of the person in beats per minute. |
| `Daily Steps` | The average number of steps the person takes per day. |
| `Sleep Disorder` | The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea). |

### Occupational Analysis

```python
# Occupational analysis: grouped box plot to illustrate sleep duration or quality across different occupations.

import matplotlib.pyplot as plt
import seaborn as sns

# Set up the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Occupation', y='Sleep Duration', data=raw_data)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.title('Occupation vs. Sleep Duration')
plt.xlabel('Occupation')
plt.ylabel('Sleep Duration')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Occupational analysis: bar plot to illustrate sleep duration or quality across different occupations.

import matplotlib.pyplot as plt

# Calculate the average sleep duration for each occupation
avg_sleep_duration = raw_data.groupby('Occupation')['Sleep Duration'].mean().sort_values()

# Set up the bar plot
plt.figure(figsize=(10, 6))
avg_sleep_duration.plot(kind='bar', color='skyblue')
plt.title('Average Sleep Duration by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Average Sleep Duration')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
```

### Sleep Duration and Quality

```python
# Scatter plots to visualize the relationship between sleep duration and quality

import matplotlib.pyplot as plt

# Set up the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x='Sleep Duration', y='Quality of Sleep', data=raw_data)
plt.title('Sleep Duration vs. Sleep Quality')
plt.xlabel('Sleep Duration')
plt.ylabel('Quality of Sleep')
plt.tight_layout()
```

### Stress Level and Sleep Quality

```python
# violin plots showing sleep quality based on stress levels

import matplotlib.pyplot as plt
import seaborn as sns

# Set up the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Stress Level', y='Quality of Sleep', data=raw_data)
plt.title('Stress Level vs. Sleep Quality')
plt.xlabel('Stress Level')
plt.ylabel('Quality of Sleep')
plt.tight_layout()
```

### BMI Category and Health Metrics

```python
# Radar Chart for BMI vs Sleep Duration
import matplotlib.pyplot as plt
import numpy as np

# Mapping BMI categories to numeric values
bmi_mapping = {'Underweight': 0, 'Normal': 1, 'Overweight': 2, 'Obese': 3}

# Convert 'BMI Category' to numeric values
raw_data['BMI Category Numeric'] = raw_data['BMI Category'].map(bmi_mapping)

# Calculate mean sleep duration for each BMI category
grouped = raw_data.groupby('BMI Category Numeric')['Sleep Duration'].mean().reset_index()

# Categories and corresponding sleep duration
categories = grouped['BMI Category Numeric']
sleep_duration = grouped['Sleep Duration']

# Labels for the radar chart
labels = ['Underweight', 'Normal', 'Overweight', 'Obese']

# Number of categories
num_categories = len(labels)

# Create a figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Set the angle for the radar chart
angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False).tolist()
angles += angles[:1]

# Fill the radar chart with sleep duration values
sleep_duration = list(sleep_duration)  # Convert to list for indexing

# Handling dimension mismatch by repeating values for each category
sleep_duration += sleep_duration[:2]

# Plotting the radar chart
ax = plt.subplot(111, polar=True)
ax.fill(angles, sleep_duration, color='skyblue', alpha=0.4)
ax.plot(angles, sleep_duration, color='blue', linewidth=2, linestyle='solid')

# Add labels for each category
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Show the plot
plt.title('Sleep Duration across BMI Categories')
plt.show()
```

```python
# line chart for heart rate vs sleep quality

import matplotlib.pyplot as plt

# Assuming df is your DataFrame with 'Heart Rate' and 'Sleep Duration' columns

# Sorting values by 'Sleep Duration' for better visualization
df = raw_data.sort_values(by='Sleep Duration')

# Set up the line plot
plt.figure(figsize=(10, 6))
plt.plot(df['Sleep Duration'], df['Heart Rate'], marker='o', linestyle='-')
plt.title('Heart Rate vs Sleep Duration')
plt.xlabel('Sleep Duration')
plt.ylabel('Heart Rate')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
